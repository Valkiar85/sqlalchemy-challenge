```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
engine.connect()
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found

inspector = inspect(engine)
inspector.get_table_names()
```

```python
type(Base.classes)
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Columns

columns = inspector.get_columns("Measurement")

for column in columns:
    print(column["name"], column["type"])
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
recent_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
print(f"The most recent date in the data set is {recent_date}")
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 

# Calculate the date one year from the last date in data set.

base_date = dt.date(2017, 8, 23) - dt.timedelta(days=365)
print(f"The date on year from the last date in the data set is {base_date}")
```

```python
# Perform a query to retrieve the data and precipitation scores

data_prcp = session.query(Measurement.date, Measurement.prcp).\
            filter(Measurement.date >= base_date).order_by(Measurement.date).all()

# Save the query results as a Pandas DataFrame and set the index to the date column

data_prcp_df = pd.DataFrame(data_prcp)
data_prcp_date_index = data_prcp_df.set_index('date')

# Sort the dataframe by date

data_prcp_sorted_df = data_prcp_date_index.sort_values(by='date')

# Use Pandas Plotting with Matplotlib to plot the data

data_prcp_sorted_df.plot(figsize=(20,10))
plt.title("Holiday in Honolulu Hawaii")
plt.ylabel("Inches")
plt.xlabel("Date")
plt.grid()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
data_prcp_sorted_df.describe()
```

# Exploratory Station Analysis

```python
# Columns Station

columns = inspector.get_columns("Station")

for column in columns:
    print(column["name"], column["type"])
```

```python
# Design a query to calculate the total number stations in the dataset

number_stations = session.query(func.count(Station.id)).all()
print(f"There are {number_stations} total number stations")
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.

most_active_stations = session.query(Measurement.station,func.count(Measurement.station)).\
group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()
print("The following are the most active stations")
most_active_stations
```

```python
# Max Station
max_station = most_active_stations[0][0]
print("The following is the max most active station")
max_station
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
lowest_temperature = session.query(func.min(Measurement.tobs)).\
filter(Measurement.station == max_station).all()

print(f"The lowest temperature is {lowest_temperature}")
```

```python
highest_temperature = session.query(func.max(Measurement.tobs)).\
filter(Measurement.station == max_station).all()
print(f"The highest temperature is {highest_temperature}")
```

```python
average_temperature = session.query(func.avg(Measurement.tobs)).\
filter(Measurement.station == max_station).all()
print(f"The average temperature is {average_temperature}")
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

temperature_observation = session.query(Measurement.date, Measurement.tobs).\
filter(Measurement.station == max_station).filter(Measurement.date > base_date).all()

temperature_observation_df = pd.DataFrame(temperature_observation)


temperature_observation_df.plot.hist(bins=12, figsize=(10,8))
plt.xlabel("Temperature")
plt.legend()
plt.grid
```
